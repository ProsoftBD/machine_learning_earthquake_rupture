### Error ananlysis on test data

#### 1. Identify true positive, negative, false positive and negative

```python
def get_classified_classes(xdata, ydata, ypred):
    true_positive = []
    true_negative = []
    false_positive = []
    false_negative = []

#     yval_new = list(yval)

    for i in range(len(ydata)):
        aa = list(xdata[i,:])
        aa.append(ydata[i])

        if ydata[i] == 0 and ypred[i] == 1:
            false_positive.append(np.array(aa))
        elif ydata[i] == 1 and ypred[i] == 0:
            false_negative.append(np.array(aa))
        elif ydata[i] == 0 and ypred[i] == 0:
            true_negative.append(np.array(aa))
        elif ydata[i] == 1 and ypred[i] == 1:
            true_positive.append(np.array(aa))

    df_true_positive = pd.DataFrame(true_positive, columns = coloumns)
    df_true_negative = pd.DataFrame(true_negative, columns = coloumns)
    df_false_positive = pd.DataFrame(false_positive, columns = coloumns)
    df_false_negative = pd.DataFrame(false_negative, columns = coloumns)
    
    df_true_positive['label'] = 1
    df_true_negative['label'] = 2
    df_false_positive['label'] = 3
    df_false_negative['label'] = 4
    
    frames = [df_true_positive, df_true_negative, df_false_positive, df_false_negative]
    df_all = pd.concat(frames)
    
    return df_all
```

```python
def discrete_cmap(N, base_cmap=None):
    base = plt.cm.get_cmap(base_cmap)
    color_list = base(np.linspace(0, 1, N))
    cmap_name = base.name + str(N)
    return base.from_list(cmap_name, color_list, N)

plt.scatter(arr[:, 0], arr[:, 1], c = arr[:, 2], cmap = discrete_cmap(4, 'rainbow'))
cbar = plt.colorbar(ticks=[1, 2, 3, 4])
cbar.ax.set_yticklabels(['True positive', 'True negative', 'False positive', 'False negative'])

plt.show()
```
